```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
plt.style.use('seaborn')
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates, y, linestyle='solid')
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate() # x-axis tick
plt.show()
```

```python
# date_format using mpl_dates.DateFormatter()

plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b %d, %Y') # Month Day, Year

plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()

plt.show()
```

```python
data = pd.read_csv('data8.csv')

price_date = data['Date']
price_close = data['Close']
```

```python
plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()


plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

```python
# transform date to pd datetime format, and sort the date with right order
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
```

```python
plt.plot_date(data['Date'], data['Close'], linestyle='solid')

plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b %d') # Month Day

plt.gca().xaxis.set_major_formatter(date_format)

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
